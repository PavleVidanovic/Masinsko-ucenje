```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("happyscore_income.csv")
df
```

```python
df.drop(['country.1'],axis=1,inplace=True)
df['region'] = df['region'].str.replace("'",'')
df
```

```python
df_prep = df.set_index('country')
df_prep
```

```python
df_scalable_noregion = df_prep.drop(['region'],axis=1)
df_scalable_noregion
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cols = df_scalable_noregion.columns
df_Scaled = scaler.fit_transform(df_scalable_noregion)
df_Scaled = pd.DataFrame(df_Scaled,columns=cols)
df_Scaled
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
relevant_columns = ['avg_satisfaction','avg_income','income_inequality','happyScore','GDP']
```

```python
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]

for num_clusters in range_n_clusters:
    
    kmeans = KMeans(n_clusters=num_clusters, max_iter=100)
    kmeans.fit(df_Scaled[relevant_columns])
    
    cluster_labels = kmeans.labels_
    
    silhouette_avg = silhouette_score(df_Scaled, cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))
```

```python
kmeans = KMeans(n_clusters=4,max_iter=100)
kmeans.fit(df_Scaled[relevant_columns])
```

```python
df_Scaled['ID'] = kmeans.labels_
df_Scaled = df_Scaled.set_index(df['country'])
df_Scaled
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x='happyScore',y='GDP',data=df_Scaled,hue='ID',palette='deep')
for i in range(df_Scaled.shape[0]):
    plt.text(x=df_Scaled.happyScore[i]-0.1,y=df_Scaled.GDP[i]+0.1,s=df_Scaled.index[i],
            fontdict=dict(size=12))
plt.title('Happiness by Country')
plt.xlabel('Happiness')
plt.ylabel('GDP')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x='happyScore',y='avg_income',data=df_Scaled,hue='ID',palette='deep')
for i in range(df_Scaled.shape[0]):
    plt.text(x=df_Scaled.happyScore[i]-0.1,y=df_Scaled.avg_income[i]+0.1,s=df_Scaled.index[i],
            fontdict=dict(size=12))
plt.title('Happiness by Country')
plt.xlabel('Happiness')
plt.ylabel('Average Income')
plt.show()
```
