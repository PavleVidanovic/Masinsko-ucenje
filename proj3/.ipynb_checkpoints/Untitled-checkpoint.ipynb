```python
import pandas as pd

df = pd.read_csv('water_potability.csv')
df
```

```python
print(df.isna().sum())
```

```python
df = df.dropna()
```

```python
df
```

```python
df.reset_index(inplace=True)
df = df.drop(['index'],axis=1)
df
```

```python
y = df['Potability']
X = df.drop(['Potability'],axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
dfx = pd.DataFrame(X)
dfx
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1337)
print(len(X_train))
print(len(X_test))
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stopping = EarlyStopping(monitor="val_loss", mode = 'auto', patience=15)
```

```python
#model = keras.Sequential([
#    layers.Dense(units=1, input_shape=[9])
#])

model = keras.Sequential()
model.add(layers.Dense(20, activation='relu', input_dim=9))
model.add(layers.Dense(1))
```

```python
opt = keras.optimizers.Adam()

model.compile(loss='mean_squared_error', optimizer=opt,metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, batch_size=75, epochs=300, 
                    validation_data=(X_test, y_test), callbacks=[early_stopping])
```

```python
import matplotlib.pyplot as plt

# Plotting the Model Accuracy & Model Loss vs Epochs (Hidden Input)
plt.figure(figsize=[20,8])

# summarize history for accuracy
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy', size=25, pad=20)
plt.ylabel('Accuracy', size=15)
plt.xlabel('Epoch', size=15)
plt.legend(['train', 'test'], loc='upper left')
```

```python
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss', size=25, pad=20)
plt.ylabel('Loss', size=15)
plt.xlabel('Epoch', size=15)
plt.legend(['train', 'test'], loc='upper left')
```
